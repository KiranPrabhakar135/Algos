## Max using recurssion and divide & conq 

```python
def max(arr):
    if len(arr) == 1:
        return arr[0]
    max_temp = max(arr[1:])
    if max_temp < arr[0]:
        return arr[0]
    else:
        return max_temp
```

```python
def max_div_and_conq(arr):
    if len(arr) == 1:
        return arr[0]
    max_left = max_div_and_conq(arr[0: floor(len(arr)/2)])
    max_right = max_div_and_conq(arr[floor(len(arr)/2):])
    return max_left if max_left > max_right else max_right
```

## Maximum sum sub array

```python
def max_sum(arr):
    print(arr)
    if len(arr) == 0:
        return None
    
    left = max_sum(arr[:-1])
    right = max_sum(arr[1:])
    center = max_sum(arr[1:-1])
    sum_all = 0
    for i in arr:
        sum_all += i
    return max([i for i in [left, right, center, sum_all] if i is not None ])
    
arr = [randint(0,100) for _ in range(0,10)]
print(str(max_sum(arr)))
```

## Path from start to end in a grid

```python
def hasPath(arr, i, j, mem):
    # print(str(i) + "*" + str(j))
    if arr[i][j] == 0:
        return []
    if len(mem[i][j]) > 0:
        return mem[i][j]
    
    res = []
    if i+1 < len(arr):
        down = hasPath(arr, i+1, j, mem)
        if len(down) > 0:
            d_res = []
            for item in down:
                d_res.append('D' + item)
            res.extend(d_res)
    if j+1 < len(arr):
        right = hasPath(arr, i, j+1, mem)
        # print('right' + str(right))
        if len(right) > 0:
            r_res = []
            for item in right:
                r_res.append('R' + item)
            res.extend(r_res)
    mem[i][j] = res
    if len(res) == 0:
        arr[i][j] = 0
    return mem[i][j]
            
rows = 4
cols = 4
arr = [[1, 0, 0, 0],[1, 1, 0, 1], [1, 1, 0, 0], [0, 1, 1, 1]]
mem = [[[] for i in range(cols)] for j in range(rows)]
mem[rows-1][cols-1] = ['C']
mem[0][0] = hasPath(arr, 0, 0, mem)
for path in mem[0][0]:
    print(path[:-1])
```

## Find 'k' largest element in stream of integers.
Constraints:
1) k can vary for every query
2) Stream can not be stored


```python
def min_heapify_insert(heap, index):
    if index == 0:
        return
    parent = int((index-1)/2)
    if heap[parent] > heap[index]:
        heap[parent], heap[index] = heap[index], heap[parent]
        min_heapify_insert(heap, parent)

def min_heapify_delete(heap, index):
    root = index
    left = 2*root + 1
    right = 2*root + 2
    
    if left < len(heap) and heap[root] > heap[left]:
        root = left
    if right < len(heap) and heap[root] > heap[right]:
        root = right
    
    if root != index:
        heap[root], heap[index] = heap[index], heap[root]
        min_heapify_delete(heap, root)

from random import randint
arr = [randint(1,100) for _ in range(20)]

# arr = [755, 400, 933, 61 ]
print(arr)
k = 4
result = [-1 for _ in range(0,k-1)]
heap = []
heap = arr[0:k]
i = int((len(heap)+1) / 2)
while i >= 0:
    min_heapify_delete(heap, i)
    i -= 1
print('The heap is: ' + str(heap))
result.append(heap[0])
for ele in arr[k:]:
    if ele > heap[0]:
        heap[0] = ele
        min_heapify_delete(heap, 0)
    result.append(heap[0]) 

print(result)
```
